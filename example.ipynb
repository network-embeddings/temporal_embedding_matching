```python
import numpy as np
from align_function import align_two_embeds
```

```python
angles = [k * np.pi/6 for k in range(12)]
shifted_angles = [k * np.pi/6 + np.pi/18 for k in range(12)]
embedding_at_t_1 = {k: [np.cos(angle), np.sin(angle)] for k,angle in enumerate(angles)}
embedding_at_t_2 = {k: [np.cos(angle), np.sin(angle)] for k, angle in enumerate(shifted_angles)}
```

```python
import matplotlib.pyplot as plt
xs = [x[0] for x in embedding_at_t_1.values()]
ys = [x[1] for x in embedding_at_t_1.values()]
p1 = plt.scatter(xs,ys,color='blue')
xs = [x[0] for x in embedding_at_t_2.values()]
ys = [x[1] for x in embedding_at_t_2.values()]
p2 = plt.scatter(xs,ys, color='red')
plt.legend((p1,p2),('embedding at t1', 'embedding at t2'),scatterpoints=1, loc=2, bbox_to_anchor=(1, 0.25))
plt.show()
```

```python
aligned_embedding_at_t_1 = align_two_embeds(embedding_at_t_1, embedding_at_t_2)
```

```python
import matplotlib.pyplot as plt
xs = [x[0] for x in embedding_at_t_1.values()]
ys = [x[1] for x in embedding_at_t_1.values()]
p1 = plt.scatter(xs,ys,color='blue')
xs = [x[0] for x in embedding_at_t_2.values()]
ys = [x[1] for x in embedding_at_t_2.values()]
p2 = plt.scatter(xs,ys, color='red')
xs = [x[0] for x in aligned_embedding_at_t_1.values()]
ys = [x[1] for x in aligned_embedding_at_t_1.values()]
p3 = plt.scatter(xs,ys, color='blue', marker='x')
plt.legend((p1,p2,p3),('embedding at t1', 'embedding at t2', 'algined embedding at t1'),scatterpoints=1, loc=2, bbox_to_anchor=(1, 0.3))
plt.show()
```
